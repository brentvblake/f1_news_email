```python
import fastf1 as ff1
import fastf1.plotting as ff1plot
from timple.timedelta import strftimedelta
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
#set cache directory
ff1.Cache.enable_cache('cache')
```

```python
year_schedule = ff1.get_event_schedule(2024)
#print column names
print(year_schedule.columns)
year_schedule
```

```python
events_remaining = ff1.get_events_remaining()
print(events_remaining.columns)
events_remaining
```

```python
def get_next_session(event_row):
    #get time of next session by checking Session#DateUtc from 1-5
    for i in range(1,6):
        closest_time = None
        session_time = event_row[f'Session{i}DateUtc']
        session_name = ''
        timestamp_now = datetime.datetime.now()
        if session_time != None or session_time != 'NaT':
            print(type(session_time))
            if session_time > timestamp_now:
                closest_time = session_time
                session_name = event_row[f'Session{i}']
                event_name = event_row['EventName']
                year = event_row['EventDate'].year
                break
            
            
    return year, event_name, session_name, closest_time

next_session = get_next_session(events_remaining.iloc[0])
print(next_session)
```

```python
session_data = ff1.get_testing_session(2024, 1, 1)
try:
    session_data.load()
except:
    print('Session not found')
```

```python
session_data.event
```

```python
session_data.laps
```

```python
drivers = pd.unique(session_data.laps['Driver'])
print(drivers)
```

```python
list_fastest_laps = []
#print columns of laps
print(session_data.laps.columns)
for driver in drivers:
    #for each tire, get fastest lap
    for tyre in pd.unique(session_data.laps['Compound']):
        fastest_lap = session_data.laps.pick_driver(driver).pick_tyre(tyre).pick_fastest()
        #if fastest lap exists, append to list
        if not pd.isna(fastest_lap['Time']):
            list_fastest_laps.append(fastest_lap)

fastest_laps = ff1.core.Laps(list_fastest_laps).sort_values('LapTime').reset_index(drop=True)

pole_lap = fastest_laps.pick_fastest()

fastest_laps['LapTimeDelta'] = fastest_laps['LapTime'] - pole_lap['LapTime']

fast_laps = fastest_laps[['Driver', 'LapTimeDelta', 'LapTime', 'Compound', 'LapNumber', 'TyreLife', 'LapStartDate']]
```

```python
#create new columns for string representation of timedelta and lap time
# fast_laps['LapTimeStr'] = strftimedelta(fast_laps['LapTime'], '%m:%s.%ms')
# fast_laps['LapTimeDeltaStr'] = strftimedelta(fast_laps['LapTimeDelta'], '%m:%s.%ms')

# #convert laptimes and lap time deltas to int of milliseconds
fast_laps['LapTimeMS'] = fast_laps['LapTime'].dt.total_seconds() * 1000
fast_laps['LapTimeDeltaMS'] = fast_laps['LapTimeDelta'].dt.total_seconds() * 1000

fast_laps
```

```python
team_colors = list()
for index, lap in fastest_laps.iterlaps():
    color = ff1plot.team_color(lap['Team'])
    team_colors.append(color)
    
fig, ax = plt.subplots()
ax.barh(fastest_laps.index, fastest_laps['LapTimeDelta'].dt.total_seconds() * 1000,
        color=team_colors, edgecolor='black')
ax.set_yticks(fastest_laps.index)
ax.set_yticklabels(fastest_laps['Driver'])
#change background color to grey of everything
ax.set_facecolor('grey')
#change colour of the outside of the plot
fig.patch.set_facecolor('grey')

ax.set_xlabel('Lap Time Delta (ms)')

# write the lap times on the bars
for i, lap in fastest_laps.iterlaps():
    
    lap_delta_string = strftimedelta(lap['LapTimeDelta'], '%s.%ms')
    compound_string = lap['Compound'][0]
    if compound_string == 'S':
        color = 'red'
    elif compound_string == 'M':
        color = 'yellow'
    elif compound_string == 'H':
        color = 'black'
    if i != 0:
        #put in white boxes
        ax.text((lap['LapTimeDelta'].total_seconds() * 1000)+100, i, lap_delta_string,
                ha='left', va='center', fontdict={'weight': 'bold'}, color='black', bbox=dict(facecolor='white', alpha=1,pad=0,edgecolor='white'))
    #also write the first letter of compound with a black stroke
    ax.text(-800, i, compound_string,
            ha='right', va='center', fontdict={'weight': 'bold'}, color=color, )
    #also write the amount of laps done
    ax.text(-1200, i, f"Lap {int(lap['LapNumber'])}",
            ha='right', va='center', fontdict={'weight': 'bold'}, color='black')
# show fastest at the top
ax.invert_yaxis()

# draw vertical lines behind the bars
ax.set_axisbelow(True)
ax.xaxis.grid(True, which='major', linestyle='--', color='black', zorder=-1000)

lap_time_string = strftimedelta(pole_lap['LapTime'], '%m:%s.%ms')

plt.suptitle(f"{session_data.event['EventName']} {session_data.event.year}\n"f"Fastest Lap: {lap_time_string} ({pole_lap['Driver']})")

plt.show()
```

```python
#convert all lap times columns to string
fast_laps['LapTime'] = fast_laps['LapTime'].strftimedelta('%M:%S.%f')
fast_laps['LapTimeDelta'] = fast_laps['LapTimeDelta'].strftimedelta('%M:%S.%f')
# make a csv file with the fastest laps
fast_laps.to_csv('fastest_laps.csv', index=False)

```
